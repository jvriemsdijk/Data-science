## Problem 1

In 2002 and 2003 Oakland Athletics, a team that used data science, had a competitive advantage. Since, others teams have started using data science as well. Use exploratory data analysis to determine if the competitive advantage has since disappeared. 

#### Problem 1(a) 
Load in [these CSV files](http://seanlahman.com/files/database/lahman-csv_2014-02-14.zip) from the [Sean Lahman's Baseball Database](http://seanlahman.com/baseball-archive/statistics). For this assignment, we will use the 'Salaries.csv' and 'Teams.csv' tables. Read these tables into a pandas `DataFrame` and show the head of each table. 

**Hint** Use the [requests](http://docs.python-requests.org/en/latest/), [StringIO](http://docs.python.org/2/library/stringio.html) and [zipfile](https://docs.python.org/2/library/zipfile.html) modules to get from the web.  

```python
def getZIP(zipFile): # Get zipfile function
    request = requests.get(zipFile).content # Requests content of zipfiles
    string = StringIO.StringIO(request)
    zf = zipfile.ZipFile(string, 'r') # Gets a list of zipfiles
    return zf

zipUrl = 'http://seanlahman.com/files/database/lahman-csv_2014-02-14.zip'
zf = getZIP(zipUrl)


tables = zf.namelist() # Return a list of file names in the archive.

# Unzip specific zip files
salaries = pd.read_csv(zf.open(tables[tables.index('Salaries.csv')]))
teams = pd.read_csv(zf.open(tables[tables.index('Teams.csv')]))

# Print heads
print "Head of player salaries:"
salaries.head()
```

```python
print "Head of teams:"
teams.head()
```

#### Problem 1(b)

Summarize the Salaries DataFrame to show the total salaries for each team for each year. Show the head of the new summarized DataFrame. 

```python
totalSalaries = salaries.groupby(['yearID','teamID'], as_index=False).sum() # as_index=False makes sure problem1(c) works.
totalSalaries.head()
```

#### Problem 1(c)

Merge the new summarized Salaries DataFrame and Teams DataFrame together to create a new DataFrame
showing wins and total salaries for each team for each year year. Show the head of the new merged DataFrame.

**Hint**: Merge the DataFrames using `teamID` and `yearID`.

```python
merged = pd.merge(totalSalaries, teams, how="inner", on=['teamID', 'yearID'])
merged.head()
```

```python
# Only project wins, total salaries for each team and for each year
winsAndSalaryDT = merged[['yearID', 'teamID', 'W', 'salary']]

winsAndSalaryDT.head()
```

#### Problem 1(d)

How would you graphically display the relationship between total wins and total salaries for a given year? What kind of plot would be best? Choose a plot to show this relationship and specifically annotate the Oakland baseball team on the on the plot. Show this plot across multiple years. In which years can you detect a competitive advantage from the Oakland baseball team of using data science? When did this end? 

**Hints**: Use a `for` loop to consider multiple years. Use the `teamID` (three letter representation of the team name) to save space on the plot.  

> To display the relationship between total wins and the total salaries in a given year we would, and will, use a scatterplot.

```python
for year in set(winsAndSalaryDT.yearID):
    
    yearlyWinsAndSalary = winsAndSalaryDT[winsAndSalaryDT.yearID == year]

    oak = yearlyWinsAndSalary[yearlyWinsAndSalary['teamID'] == 'OAK']

    
    yearlyWinsAndSalary.plot(x='W', y='salary', use_index=False, title=year, kind='scatter').annotate(
                    'OAK', xy=(oak.W, oak.salary), xycoords='data',
                    xytext=(0.8, 0.95), textcoords='axes fraction',
                    arrowprops=dict(facecolor='black', shrink=0.05),
                    horizontalalignment='right', verticalalignment='top')


```

In several years the Oakland A's have been performing better than their total salary expenses would indicate. These years are:

- 2013

With the biggest advantage clear at 2002, tied for the most wins, but at a fraction of the salary costs of their counterpart

#### Problem 1(e):

Fit a linear regression to the data from each year and obtain the residuals. Plot the residuals against time to detect patterns that support your answer in 1(d). 

```python
# we only look at the years 1999 to 2005 because they are the most relevant to the question
years = np.arange(1999, 2005)
residuals = pd.DataFrame()

for year in years: 
    df = merged[merged['yearID'] == year]
    xAxis = df['salary'].values # Values in salary
    xAxis = xAxis / 1000000 # Divided by 1000000 for a better visible plotting
    yAxis = df['W'].values # Values in Wins.

    # Estimate of the least squares
    wins = np.array([xAxis, np.ones(len(xAxis))])
    
    # numpy.linalg.lstsq returns least squares solutions to a linear matrix equation
    leastSquares = np.linalg.lstsq(wins.T, yAxis)[0] # coefficients through numpy
    regLine = (leastSquares[0] * xAxis + leastSquares[1]) # regression line
    # Wins minus the regression line
    residuals[year] = yAxis - regLine
  
residuals.index = df['teamID']
residuals = residuals.T # Transpose residuals

# Oakland baseball team is displayed in red, rest in gray
residuals.plot(title = 'linear regression', color=map(lambda teamID: 'red' if teamID=='OAK' else 'gray',df.teamID))

print plt.show()
```
