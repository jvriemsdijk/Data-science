## 1

```python
def analyzeLang():
    """
    Method to read the unesco lang-lang translation file and build
    our own user-lang dataframe.
    """

    # open number of translation data as a dataframe
    langTrans = pd.DataFrame.from_csv('unesco_langlang_20120722_iso639-3.txt', sep='\t')

    # setup a dict to generate a dataframe with later
    langDict = {}
    langSet = set(langTrans.index.values) | set(langTrans.target_language.values)
    for lang in langSet:
        langDict[lang] = {'translated_from': 0, 'translated_to': 0, 'out_degree': 0, 'in_degree': 0}

    # loop over all translation numbers    
    for index, row in langTrans.iterrows():
        langDict[index]['translated_from'] += row['num_translations']
        langDict[index]['out_degree'] += 1
        langDict[row['target_language']]['translated_to'] += row['num_translations']
        langDict[row['target_language']]['in_degree'] += 1


    langDf = pd.DataFrame.from_dict(langDict)
    langDf = langDf.transpose()
    langDf.index.name = 'langcode'
    
    return langDf

langDf = analyzeLang()

langDf.head()
```

## 1

```python
langGraph = nx.DiGraph()
langTrans = pd.DataFrame.from_csv('unesco_langlang_20120722_iso639-3.txt', sep='\t')

for index, row in langTrans.iterrows():
    langGraph.add_edge(index, row['target_language'], weight = row['num_translations'])
```

## 2

```python
def drawGraph(graph, random=False):
    
    if random:
        
        pos=nx.random_layout(graph) # positions for all nodes

    else :
        
        pos=nx.spring_layout(graph) # positions for all nodes
    # nodes
    nx.draw_networkx_nodes(graph, pos, node_size=500)

    # edges
    nx.draw_networkx_edges(graph, pos,
                        width=1)

    # labels
    nx.draw_networkx_labels(graph, pos, font_size=10, font_family='sans-serif')

    plt.axis('off')
    # plt.savefig("weighted_graph.png") # save as png
    plt.show() # display
    
```

```python
drawGraph(langGraph)
```

This is ofcourse not readable, so I will remake the graph with just the strongly connected languages.

```python

for i in nx.strongly_connected_component_subgraphs(langGraph):
    if len(i.edges()) > 0:
        strongGraph = i
  

drawGraph(strongGraph)
```
