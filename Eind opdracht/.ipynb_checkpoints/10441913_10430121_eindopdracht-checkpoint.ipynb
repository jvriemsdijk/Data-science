##Abstract


Dit explorerende onderzoek is gericht op het uitwijzen of naties oorlogszuchtig te noemen zijn, op basis van offensieve deelname aan veldslagen. Middels de CDB13 dataset wordt een algoritme getest dat een basis kan vormen voor een vervolgstudie. Tevens worden op basis van de resultaten die uit dit algoritme volgen visualisaties gegenereerd. Aan de hand van deze visualisaties kunnen uitspraken gedaan worden over het verloop van specifieke oorlogen en de trend die naties volgen door de oorlogen heen. #Introduction

Door de loop der eeuwen is een groot aantal veldslagen uitgevochten. Bestaansrecht, macht of politieke spanningen zijn allen meerdere malen de oorzaak geweest van een treffen op het slagveld. Wetenschappers zijn, ondanks deze puur menselijke vorm van interactie, nog niet eens wat de oorzaak is van het willen strijden tegen elkaar.

Biologen zijn ervan overtuigd dat de mens 'wired for war' is. De mens is geevolueerd tot vechtmachine die dankzij zijn vermogen tot conceptueel denken, zijn uithoudingsvermogen, capaciteit om binoculair te kijken en goede coordinatie zich onderscheid van dieren. Psychologen zijn er echter van overtuigd dat de mens beinvloedbaar is. Oorlog voeren zou vanzelf volgen na voldoende indoctrinatie of stimuli. Er is een drang naar spanning en actie.

Op internationale basis lijkt er ook een onderscheid gemaakt te kunnen worden. Er zijn landen die puur defensief handelen, of uberhaupt nooit een grote veldslag hebben uitgevochten. Andere landen lijken echter continu in staat van oorlog te verkeren. Von der Goltz was overtuigd dat oorlog voor naties een onvermijdelijk lot is (1899). Hier sluit bij aan dat oorlog vaak gezien wordt als continuatie is van de politiek en diplomatie (Von Clausewitz, 2004). Het leger is dan ook een instrument om ingezet te worden door de politieke machthebbers. Volgens deze redenatie zou het puur afhankelijk van het politieke cultuur zijn of een natie dus gedefinieerd zou kunnen worden als oorlogszuchtig of pacifistisch.

Er zullen naties zijn die veel meer vechten dan andere landen. De vraag is dan echter of deze landen dan niet in een gelijk aantal veldslagen defensief moeten handelen. In dit onderzoek zal aan de hand van de volgende onderzoeksvraag een antwoord worden gezocht op de vraag of er specifieke landen zijn die continu oorlogen beginnen:

## Methodology
De gebruikte dataset in dit onderzoek is een verbeterde versie van de CDB13 database uit 1993, die opzich weer is opgebouwd uit data uit de CDB90 dataset. Deze bestaat uit een set van 660 veldslagen tussen 1600 en 1975. Naast de algemene data als de veldslag en de strijdende partijen, wordt data over onder andere het weer, mankracht en tactische omgevingsfactoren meegeleverd. Specifieker zijn dit bijvoorbeeld de breedte van het front en het aantal cavaleristen. Voor dit onderzoek maken wij gebruik van de 53 kolommen in de battles.csv, battleActors.csv en battleDurations.csv sets. Zoals de bestandsnamen al aanduiden, zijn dit sets bestaande uit respectievelijk de veldslagen, deelnemende landen en de duur van de betreffende veldslagen. 

De data is oorspronkelijk verzameld door de US ARMY in de CDB90 dataset. Deze is vervolgens omgezet en opgeschoond in de CDB13 set. Voor dit onderzoek is gekozen om de CSV- of Comma Separated Value-versie te gebruiken, aangezien deze direct vanaf Github te downloaden was. Als alternatief valt ook handmatig elke tabel als JSON-bestand te downloaden.

De data wordt in zip-vorm ingeladen vanuit Jeffrey Arnold's [Github repository][1]. Deze docent aan de universiteit van Washington is mede verantwoordelijk geweest voor het om zetten van de CDB90-set in de CDB13-set. Na het inladen worden vervolgens geautomatiseerd door Python-code de bestanden uitgelezen te worden. De tabellen zijn in CSV-formaat, wat wil zeggen dat de data op basis van komma's gesplitst is. Middels de functie pd.read_csv zullen deze CSV's vervolgens door een Python-package genaamd Pandas ingelezen worden in een Dataframe.

Met een analyse van de data wordt getracht de onderzoeksvraag te beantwoorden en de hypotheses aan te nemen dan wel te verwerpen. Er wordt gekeken in hoeverre naties oorlogen starten en offensief in deze oorlogen blijven handelen. Dit wordt gedaan aan de hand van het aantal veldslagen dat als aanvallende partij wordt gestart. Deze resultaten zullen vervolgens gepresenteerd worden in verschillende diagrammen en visualisaties.

Het onderzoek draait om een te geven aggressie-score per land. Allereerst wordt er gekeken of er al een score bekend is voor dit betreffende land. Mocht dit het geval zijn, dan zal voor de aggressor een halve punt toegevoegd worden aan zijn score. Eenzelfde stap kan genomen worden voor de defensieve partij, waar een halve punt in mindering gebracht wordt. Door te werken met punten aftrek en de mogelijkheid tot negatieve getallen, valt te onderzoeken of een natie minder aggresief wordt door de loop der tijd. Mogelijk valt dit zelfs te concluderen op een kleiner niveau, namelijk dat van een enkele oorlog.

```python
# imports and general setup
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import requests
import StringIO
import zipfile
import scipy.stats 
import json
import argparse
import os
import sqlite3
import csv
import codecs

import urlparse
import urllib2

from IPython.display import HTML


# sets the graphs to a larger size
plt.rcParams['figure.figsize'] = (15, 10)
```

```python
def getZIP(): # Get zipfile function
    request = requests.get('https://github.com/jrnold/CDB13/archive/master.zip').content # Requests content of zipfiles
    string = StringIO.StringIO(request)
    zf = zipfile.ZipFile(string, 'r') # Gets a list of zipfiles
    return zf
```

```python
# download the CDB13 zip from github
zippedDB = getZIP()
dbFiles = zippedDB.namelist()
```

```python
battles = pd.read_csv(zippedDB.open(dbFiles[dbFiles.index('CDB13-master/data/battles.csv')]), index_col='isqno')
battleActors = pd.read_csv(zippedDB.open(dbFiles[dbFiles.index('CDB13-master/data/battle_actors.csv')]))
# battleDurations = pd.read_csv(zippedDB.open(dbFiles[dbFiles.index('CDB13-master/data/battle_durations.csv')]), index_col='isqno', parse_dates=['datetime', 'datetime_min', 'datetime_max'])
battleDurations = pd.read_csv(zippedDB.open(dbFiles[dbFiles.index('CDB13-master/data/battle_durations.csv')]), index_col='isqno')

battles = pd.merge(battles, battleDurations, right_index=True, left_index=True)
battles.head()
```

```python
def makeWarInfoDataFrame(warName):
    """
    Method to which returns two dataframes:
    the first one contains some statistics about all actors in the war,
    the second contains the aggressor rating over the course of the battles in the given war.
    """
    
    warAggressorDict = {}
    aggressorDict = {}
    
    columnNames = []
    
    warBattles = battles.query('war == "' + warName + '"')
    warBattles = warBattles.sort(columns = 'datetime_min', ascending=True)
    
    for battle in warBattles.iterrows():
        
        # timestamp van battle toevoegen
        aggressorDict['datetime'] = battle[1].datetime_min
        
        # actors opzoeken
        actors = battleActors.loc[battleActors['isqno'].isin([battle[0]])]
        
        # aggressor score bepalen
        for actorRow in actors.iterrows():
            # kijken of er al een score bekend is per actor
            if actorRow[1].actor in aggressorDict.keys():
                # ja, verder rekenen
                aggressorDict[actorRow[1].actor] += float(actorRow[1].attacker - 0.5)
            else:
                # nee, toevoegen
                aggressorDict[actorRow[1].actor] = float(actorRow[1].attacker - 0.5)
        
            columnNames.append(actorRow[1].actor)
        
        # toevoegen in return dict en tussenstand dict
        warAggressorDict[battle[1][1]] = aggressorDict.copy()
        
    columnNames = list(set(columnNames))
        
    aggroDF = pd.DataFrame.from_dict(warAggressorDict, orient='index')
    aggroDF = aggroDF.fillna(0)
    aggroDF = aggroDF.sort(columns=columnNames)
    aggroDF['datetime'] = pd.to_datetime(aggroDF['datetime'], coerce=True)
    
    return aggroDF.sort(columns='datetime')
```

```python
def makeTotalInfoDataFrame():
    """
    Method which returns two dataframes:
    the first one contains some statistics about all actors in the war,
    the second contains the aggressor rating over the course of the battles in the given war.
    """ 
    
    warAggressorDict = {}
    aggressorDict = {}
    columnNames = []
    
    warBattles = battles.sort(columns = 'datetime_min', ascending=True)
    
    for battle in warBattles.iterrows():
        
        
        if type(pd.to_datetime(battle[1].datetime_min, coerce=True)) != pd.tslib.NaTType:
        
            # timestamp van battle toevoegen
            aggressorDict['datetime'] = battle[1].datetime_min

            # actors opzoeken
            actors = battleActors.loc[battleActors['isqno'].isin([battle[0]])]

            # aggressor score bepalen
            for actorRow in actors.iterrows():
                # kijken of er al een score bekend is per actor
                if actorRow[1].actor in aggressorDict.keys():
                    # ja, verder rekenen
                    aggressorDict[actorRow[1].actor] += float(actorRow[1].attacker - 0.5)
                else:
                    # nee, toevoegen
                    aggressorDict[actorRow[1].actor] = float(actorRow[1].attacker - 0.5)

                columnNames.append(actorRow[1].actor)

            # toevoegen in return dict en tussenstand dict
            warAggressorDict[battle[1][1]] = aggressorDict.copy()
        
    columnNames = list(set(columnNames))

    aggroDF = pd.DataFrame.from_dict(warAggressorDict, orient='index')
    aggroDF = aggroDF.fillna(0) 
    aggroDF = aggroDF.sort(columns=columnNames)
    aggroDF['datetime'] = pd.to_datetime(aggroDF['datetime'])
    
    return aggroDF.sort(columns='datetime')
```

```python
allCountriesDF = makeTotalInfoDataFrame()
```

```python
for actorName in actorSet:
    if actorName in allCountriesDF.columns:
        allCountriesDF[['datetime', actorName]].plot(x='datetime', title=actorName)
```
