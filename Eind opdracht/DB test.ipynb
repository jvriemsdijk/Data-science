```python
def getZIP(): # Get zipfile function
    request = requests.get('https://github.com/jrnold/CDB13/archive/master.zip').content # Requests content of zipfiles
    string = StringIO.StringIO(request)
    zf = zipfile.ZipFile(string, 'r') # Gets a list of zipfiles
    return zf
```

```python
# download the CDB13 zip from github
zippedDB = getZIP()
dbFiles = zippedDB.namelist()

# for i in dbFiles:
#     if i[12:18] == ('/data/') and i[-4:] == '.csv':
#         print i
```

```python
battles = pd.read_csv(zippedDB.open(dbFiles[dbFiles.index('CDB13-master/data/battles.csv')]), index_col='isqno')
battleActors = pd.read_csv(zippedDB.open(dbFiles[dbFiles.index('CDB13-master/data/battle_actors.csv')]))
# battleDurations = pd.read_csv(zippedDB.open(dbFiles[dbFiles.index('CDB13-master/data/battle_durations.csv')]), index_col='isqno', parse_dates=['datetime', 'datetime_min', 'datetime_max'])
battleDurations = pd.read_csv(zippedDB.open(dbFiles[dbFiles.index('CDB13-master/data/battle_durations.csv')]), index_col='isqno')

battles = pd.merge(battles, battleDurations, right_index=True, left_index=True)
battles.head()
```

```python
def makeWarInfoDataFrame(warName):
    """
    Method to which returns two dataframes:
    the first one contains some statistics about all actors in the war,
    the second contains the aggressor rating over the course of the battles in the given war.
    """
    
    
    warInfoDict = {}
    warAggressorDict = {}
    aggressorDict = {}
    battleCount = 1.0
    
    columnNames = []
    
    warBattles = battles.query('war == "' + warName + '"')
    warBattles = warBattles.sort(columns = 'datetime_min', ascending=True)
    
    for battle in warBattles.iterrows():
        
        # timestamp van battle toevoegen
        aggressorDict['datetime'] = battle[1].datetime_min
        
        # actors opzoeken
        actors = battleActors.loc[battleActors['isqno'].isin([battle[0]])]
        
        # aggressor score bepalen
        for actorRow in actors.iterrows():
            # kijken of er al een score bekend is per actor
            if actorRow[1].actor in aggressorDict.keys():
                # ja, verder rekenen
                aggressorDict[actorRow[1].actor] += float(actorRow[1].attacker - 0.5)
            else:
                # nee, toevoegen
                aggressorDict[actorRow[1].actor] = float(actorRow[1].attacker - 0.5)
        
            columnNames.append(actorRow[1].actor)
        
        # toevoegen in return dict en tussenstand dict
        warAggressorDict[battle[1][1]] = aggressorDict.copy()
        
        battleCount += 1.0
        
    columnNames = list(set(columnNames))
        
    aggroDF = pd.DataFrame.from_dict(warAggressorDict, orient='index')
    aggroDF = aggroDF.fillna(0)
    aggroDF = aggroDF.sort(columns=columnNames)
    aggroDF['datetime'] = pd.to_datetime(aggroDF['datetime'], coerce=True)
    
    return aggroDF.sort(columns='datetime')
    
```

```python
# From https://github.com/okfn/dptools/blob/master/bin/load-sqlite.py
# No license given in original. Since OKFN assumed to be have open source license.
#
# Edits by Jeffrey Arnold (2014), BSD-3
#
# - converted to Python 3
#
# Further edits by Joey van Riemsdijk (2015)
#
# Now for iPython 2.2 & python 2.7



# http://www.sqlite.org/datatype3.html
mappings = {
    'string': 'text',
    'number': 'real',
    'float': 'real',
    'integer': 'integer',
    'date': 'string',
    'boolean': 'integer',
    'date': 'date',
    'datetime': 'datetime'
}


def load(dpurlOrPath, sqlitePath):
    if not dpurlOrPath.split('://')[0] in ['http', 'https', 'ftp']:
        # it's a path
        dpurl = path2url(dpurlOrPath)
    else:
        dpurl = dpurlOrPath
        # urljoin makes this work right -
        # http://docs.python.org/2/library/urlparse.html#urlparse.urljoin
        dpurl = urlparse.urljoin(dpurl, 'datapackage.json')
        basepath = dpurlOrPath.rstrip('datapackage.json')
        dpfo = urllib2.urlopen(dpurl)
        reader = codecs.getreader("utf-8")
        out = json.load(reader(dpfo))
        for finfo in out['resources']:
            # normalization so we do not have to handle both alternatives
            if not 'url' in finfo:
                finfo['url'] = urlparse.urljoin(dpurl, finfo['path'])
            process_resource(finfo, sqlitePath)


def process_resource(finfo, dbpath):
    '''Load the resource specified by finfo into the database at dbpath
    '''
    if 'name' in finfo:
        tablename = finfo['name']
    else:
        _fname = urlparse.urlparse(finfo.get('url', ''))[2].split('/')[-1]
        tablename = os.path.splitext(_fname)[0]
        
    fields = finfo['schema']['fields']
    _columns = ','.join(
        ['"%s" %s' % (field['id'], mappings[field['type']])
         for field in fields]
    )
    conn = sqlite3.connect(dbpath)
    c = conn.cursor()
    sql = 'CREATE TABLE "%s" (%s)' % (tablename, _columns)
    c.execute(sql)
    _insert_tmpl = 'insert into "%s" values (%s)' % (tablename,
                                                     ','.join(['?'] * len(fields)))
    # could do this on but not very robust and will not work on remote urls ...
    # sqlite> .mode csv
    # sqlite> .import <filename> <table>
    req = urllib2.urlopen(finfo['url'])
    reader = csv.reader(req.read().decode('utf-8').splitlines())
    # skip headers
    for row in reader:
        c.execute(_insert_tmpl, row)
    conn.commit()
    c.close()


def path2url(path):
    return urlparse.urljoin('file:', requests.get(urllib.pathname2url(os.path.abspath(path))))

# if __name__ == '__main__':
#     parser = argparse.ArgumentParser(description='Load a data package into sqlite.')
#     parser.add_argument('datapackage', help='path to datapackage')
#     parser.add_argument('sqlite', help='path to sqlite db')
#     args = parser.parse_args()
#     load(args.datapackage, args.sqlite)
```
